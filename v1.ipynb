```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv('netflix1.csv')
df.info()
```

```python
df['date_added'] = pd.to_datetime(df['date_added'])
df.info()
```

```python
df_null = df.isnull().sum()
df_null
```

```python
df.head()
```

```python
sns.countplot(x = 'type', data = df)
plt.title('Number of movies and TV-show')
plt.show()
```

```python
df_movies = df[df['type'] == 'Movie']
df_movies
```

```python
df_movies['duration'] = df_movies['duration'].apply(lambda x: int(x.replace('min','')))
df_movies
```

```python
plt.scatter(x = 'duration', y = 'release_year', data= df_movies)
plt.xlabel('Duration')
plt.ylabel('Year')
plt.title('Duration and year movies were released.')
plt.show()
```

```python
mcc = {}
mcc['Others'] = 0
movie_count_country = df_movies['country'].value_counts()
for key, value in movie_count_country.items():
    if value <= 100:
        mcc['Others'] += 1
    elif key == 'Not Given':
        mcc['Others'] += 1
    else:
        mcc[key] = value
mcc_series = pd.Series(mcc)
```

```python
mcc_series.plot(kind='pie', autopct='%1.1f%%')
plt.title('Movies from all over the world')
plt.show()
```

```python
df_movies['listed_in'] = df_movies['listed_in'].apply(lambda x: x.split(','))
df_movies
```

```python
df_movies_type = df_movies.explode('listed_in')
df_movies_after_2020 = df_movies_type[df_movies_type['release_year'] >= 2020]
```

```python
sns.countplot(x = 'listed_in', data = df_movies_after_2020, hue='release_year')
plt.xticks(None, rotation='vertical')
plt.show()
```
